# Import all 4 datasets into MongoDB

```python
# Import dependencies.
import pymongo
import json
import csv
```

```python
# Make a connection to MongoDB.
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the 'project_3' database in MongoDB.
db = client.project_3

# Define the header.
header = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 
         'genres', 'description']
```

```python
#Convert CSV to JSON for Netflix data.
csvfile = open('../csv/cleaned_data/netflix.csv', 'r', encoding="utf8")
reader = csv.DictReader(csvfile)

for each in reader:
    row = {}
    for field in header:
        row[field] = each[field]
 
    # Upload each row of data to 'netflix' collection in the DB.
    db.netflix.insert_one(row)
```

```python
#Convert CSV to JSON for Hulu data.
csvfile = open('../csv/cleaned_data/hulu.csv', 'r', encoding="utf8")
reader = csv.DictReader(csvfile)

for each in reader:
    row = {}
    for field in header:
        row[field] = each[field]
 
    # Upload each row of data to 'hulu' collection in the DB.
    db.hulu.insert_one(row)
```

```python
#Convert CSV to JSON for Disney Plus data.
csvfile = open('../csv/cleaned_data/disney_plus.csv', 'r', encoding="utf8")
reader = csv.DictReader(csvfile)

for each in reader:
    row = {}
    for field in header:
        row[field] = each[field]
 
    # Upload each row of data to 'disney_plus' collection in the DB.
    db.disney_plus.insert_one(row)
```

```python
#Convert CSV to JSON for Amazon Prime data.
csvfile = open('../csv/cleaned_data/amazon_prime.csv', 'r', encoding="utf8")
reader = csv.DictReader(csvfile)

for each in reader:
    row = {}
    for field in header:
        row[field] = each[field]
 
    # Upload each row of data to 'amazon_prime' collection in the DB.
    db.amazon_prime.insert_one(row)
```

# Rotten Tomatoes File Introduction

```python
# Save the Rotten Tomatoes CSV file path and read it into a DataFrame.
rotten_csv = '../csv/original_data/MoviesOnStreamingPlatforms.csv'

rotten_df = pd.read_csv(rotten_csv)

# Save each cleaned CSV file path into a variable.
netflix_cleaned = '../csv/cleaned_data/netflix.csv'
hulu_cleaned = '../csv/cleaned_data/hulu.csv'
disney_plus_cleaned = '../csv/cleaned_data/disney_plus.csv'
amazon_prime_cleaned = '../csv/cleaned_data/amazon_prime.csv'

# Read each cleaned CSV file into a DataFrame.
netflix_df = pd.read_csv(netflix_cleaned)
hulu_df = pd.read_csv(hulu_cleaned)
disney_plus_df = pd.read_csv(disney_plus_cleaned)
amazon_prime_df = pd.read_csv(amazon_prime_cleaned)
```

```python
# Take out '/100' from Rotten Tomatoes scores.
rotten_df['Rotten Tomatoes'] = rotten_df['Rotten Tomatoes'].str.replace('/100', '')
```

```python
# Drop unnecessary columns.
rotten_df = rotten_df.drop(columns=['Age', 'Type', 'Unnamed: 0', 'Year'])
```

```python
# Check data summary.
rotten_df.info()
```

```python
# Drop all rows that do not have Rotten Tomatoes scores.
rotten_df = rotten_df.dropna()
```

```python
# Check data summary.
rotten_df.info()
```

```python
# Preview data.
rotten_df
```

```python
# Join our four cleaned datasets into one.
combined_df = pd.concat([netflix_df, hulu_df, disney_plus_df, amazon_prime_df], ignore_index=True) 
```

```python
# Drop the 'show_id' column.
combined_df = combined_df.drop(columns=['show_id'])
```

```python
# Rename 'title' column to 'Title' for purposes of merging combined_df with rotten_df.
combined_df = combined_df.rename(columns={'title': 'Title'})
```

```python
combined_df
```

```python
# Merge combined_df with rotten_df.
merged_df = rotten_df.merge(combined_df, how='inner', on='Title')
```

```python
merged_df
```

```python
# Drop unnecessary columns.
merged_df = merged_df.drop(columns=['Netflix', 'Hulu', 'Prime Video', 'Disney+', 'ID'])
```

```python
# Check value counts for each platform.
merged_df['platform'].value_counts()
```

```python
# Change Rotten Tomatoes scores to an integer data type.
merged_df['Rotten Tomatoes'] = merged_df['Rotten Tomatoes'].astype(int)
```

```python
# Rename columns for uniformity's sake.
merged_df = merged_df.rename(columns={'Title': 'title', 'Rotten Tomatoes': 'score', 
                                      'date_added': 'added', 'release_year': 'release'})
```

```python
# Check data summary.
merged_df.info()
```

```python
# Check the movie and tv show value counts.
merged_df['type'].value_counts()
```

```python
# Drop rows of tv shows.
merged_df = merged_df.drop(merged_df[merged_df['type'] == 'TV Show'].index)
```

```python
# Check data summary.
merged_df.info()
```

```python
# Drop 'type' column.
merged_df = merged_df.drop(columns=['type'])
```

```python
# Reset the index.
merged_df = merged_df.reset_index(drop=True)
merged_df
```

```python
# Export merged_df as a CSV file.
merged_df.to_csv('../csv/cleaned_data/merged_unfinished.csv', index=False)
```

# Genre Splitting

```python
# Import dependencies.
import pandas as pd
```

```python
# Save the merged_unfinished CSV file path and read it into a DataFrame.
merged_csv = '../csv/cleaned_data/merged_unfinished.csv'

merged_data = pd.read_csv(merged_csv)
```

```python
# Preview data.
merged_data
```

```python
# Create a list of IDs to be used in our new 'ID' column.
IDs = []

for i in range(0, 6956):
    IDs.append(i+1)
    
# Create the 'ID' column and position it as the first column in the DataFrame.
merged_data.insert(0, 'ID', IDs)
```

```python
# Preview data.
merged_data
```

```python
# Clean up nomenclature for 'listed_in' column.
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Anime Features', 'Anime')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Classics', 'Classic Movies')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Comedy', 'Comedies')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Documentary', 'Documentaries')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Dramas', 'Drama')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Faith and Spirituality', 'Faith & Spirituality')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Historical', 'History')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Horror Movies', 'Horror')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('International Movies', 'International')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('+', '', regex=False)
merged_data['listed_in'] = merged_data['listed_in'].str.replace('LGBTQ Movies', 'LGBTQ')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Romantic Movies', 'Romance')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Science Fiction', 'Sci-Fi & Fantasy')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Sports Movies', 'Sports')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Stand-Up Comedies', 'Stand Up')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Thrillers', 'Thriller')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Young Adult Audience', 'Teen')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Action-Adventure', 'Action & Adventure')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('and Culture', '')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Concert Film', 'Music')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Music & Musicals', 'Music')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Music Videos and Concerts', 'Music')
merged_data['listed_in'] = merged_data['listed_in'].str.replace('Musical', 'Music')
```

```python
# Export merged_data as a CSV file.
merged_data.to_csv('../csv/cleaned_data/merged_finished.csv', index=False)
```

```python
# Split up genres in 'listed_in' column and divide them up individually in a new 'genres_df'.
def splitting (dataframe, col):
    result = dataframe[col].str.get_dummies(', ')
    return result

genres_df = splitting(merged_data, 'listed_in')
```

```python
# Create the 'ID' column and position it as the first column in the DataFrame.
genres_df.insert(0, 'ID', IDs)
```

```python
# Preview data.
genres_df
```

```python
# Drop 'type' column.
genres_df = genres_df.drop(columns=['Movies'])
```

```python
# Data summary.
genres_df.info()
```

```python
# Export 'genres_df' as a CSV file.
genres_df.to_csv('../csv/cleaned_data/genres.csv', index=False)
```

# Genre Splitting ALT

```python
# Import dependencies.
import pandas as pd

# Save both CSV file paths and read them into separate DataFrames.
merged_csv = '../csv/cleaned_data/merged_finished.csv'
merged_data = pd.read_csv(merged_csv)

genres_csv = '../csv/cleaned_data/genres.csv'
genres_data = pd.read_csv(genres_csv)
```

```python
# Preview data.
merged_data
```

```python
# Convert 'added' column into datetime format.
merged_data['added'] = merged_data['added'].str.replace('-', '')

merged_data['added'] = pd.to_datetime(merged_data['added'], format='%Y%m%d')
```

```python
# Preview data types.
merged_data.dtypes
```

## Horizontal Version

```python
# Preview data.
genres_data
```

```python
# Merge 'merged_data' and 'genres_data' together.
complete_horizontal = merged_data.merge(genres_data, on='ID')
```

```python
# Drop the 'listed_in' column since it is now obsolete because of the data from 'genres_data'.
complete_horizontal = complete_horizontal.drop(columns=['listed_in'])
complete_horizontal
```

## Vertical Version

```python
# Create a list of all genres for each movie by dividing up the strings in the 'listed_in' column.
new_list = merged_data['listed_in'].str.split(', ').to_list()
```

```python
# Drop the old 'listed_in' column.
merged_data = merged_data.drop(columns=['listed_in'])
merged_data
```

```python
# Add the new 'listed_in' column.
merged_data.insert(10, 'listed_in', new_list)
merged_data
```

```python
# Expand our DataFrame by each genre of each movie in the 'listed_in' column.
complete_vertical = merged_data.explode('listed_in').reset_index(drop=True)
complete_vertical
```

```python
# Export both complete DataFrames as CSV files.
complete_horizontal.to_csv('../csv/complete_data/complete_horizontal.csv', index=False)

complete_vertical.to_csv('../csv/complete_data/complete_vertical.csv', index=False)
```

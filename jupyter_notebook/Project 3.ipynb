```python
# Import dependencies.
import pandas as pd
```

```python
# Save each CSV file path into a variable.
netflix_csv = '../original_data/netflix_titles.csv'
hulu_csv = '../original_data/hulu_titles.csv'
disney_plus_csv = '../original_data/disney_plus_titles.csv'
amazon_prime_csv = '../original_data/amazon_prime_titles.csv'
```

```python
# Read each CSV file into a DataFrame.
netflix_data = pd.read_csv(netflix_csv)
hulu_data = pd.read_csv(hulu_csv)
disney_plus_data = pd.read_csv(disney_plus_csv)
amazon_prime_data = pd.read_csv(amazon_prime_csv)
```

```python
# Check data types.
netflix_data.dtypes
```

```python
# Preview data.
netflix_data
```

```python
# Replace month names with their respective numbers, and delete spaces and commas so that we may convert each date
# into datetime format.
netflix_data['date_added'] = netflix_data['date_added'].str.replace('January', '01')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('February', '02')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('March', '03')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('April', '04')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('May', '05')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('June', '06')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('July', '07')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('August', '08')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('September', '09')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('October', '10')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('November', '11')
netflix_data['date_added'] = netflix_data['date_added'].str.replace('December', '12')
netflix_data['date_added'] = netflix_data['date_added'].str.replace(' ', '')
netflix_data['date_added'] = netflix_data['date_added'].str.replace(',', '')
```

```python
# Convert 'date_added' column into datetime format.
netflix_data['date_added'] = pd.to_datetime(netflix_data['date_added'], format='%m%d%Y')
```

```python
netflix_data = netflix_data.rename(columns={'listed_in': 'genres'})
```

```python
# Check data summary.
netflix_data.info()
```

# Import all 4 datasets into MongoDB

```python
# Import dependencies.
import pymongo
import json
import csv
```

```python
# Make a connection to MongoDB.
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the 'project_3' database in MongoDB.
db = client.project_3

# Define the header.
header = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 
         'genres', 'description']
```

```python
#Convert CSV to JSON for Netflix data.
csvfile = open('../cleaned_data/netflix.csv', 'r', encoding="utf8")
reader = csv.DictReader(csvfile)

for each in reader:
    row = {}
    for field in header:
        row[field] = each[field]
 
    # Upload each row of data to 'netflix' collection in the DB.
    db.netflix.insert_one(row)
```
